### 4.)

```python
#creating the max_order column while splitting the data based on the user id column and transforming the order_number column for maximum orders for each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#checking for the new column
ords_prods_merge.head(15)
```

### 5.)

```python
#create a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

Basic Statistics of each category.

```python
#using groupby for 'loyalty_flag' and aggregating 'prices' aggregate functions mean min max
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

We can see that the highest mean belongs to the new customers, the lowest mean belongs to the loyal customer, and the regular customers are in between. We can conclude that there is a slight difference in whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers. New customers pay a little more than regular customers. Regular customers pay a little more than loyal customers.

### 6.)

```python
#creating the average_spending column while splitting the data based on the user id column and transforming the prices column for average orders for each user
ords_prods_merge['average_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#checking for the new column
ords_prods_merge.head()
```

```python
#create a flag that assigns a “spending” label to a user ID based on its corresponding average price value
ords_prods_merge.loc[ords_prods_merge['average_spending'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spending'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
#checking the freqency of customers in the 'spending_flag' column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

There are a majority of low spenders compared to high spenders.

### 7.)

```python
#creating the order frequency column while splitting the data based on the user id column and transforming the 'days_since_prior_order' column for median orders for each user
ords_prods_merge['order_regularity'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#checking for the new column
ords_prods_merge.head()
```

```python
#create a flag that assigns a “order_frequency” label to a user ID based on its corresponding median 'days_since_prior_order' value
ords_prods_merge.loc[ords_prods_merge['order_regularity'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_regularity'] > 10) & (ords_prods_merge['order_regularity'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_regularity'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#checking the freqency of customers in the 'order_frequency_flag' column
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

There are a majority of frequent customers than regular customers. There are more regular customers than non-frequent customers.
