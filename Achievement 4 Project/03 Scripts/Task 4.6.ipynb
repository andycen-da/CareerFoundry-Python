### 05.) Determine a suitable way to combine the orders_products_combined dataframe with your products data set.

```python
#outer join
#When performing an outer merge, it takes results from both dataframes, whether or not they match on the key column.
df_ords_prods_merge_left = df_ords_products_combined.merge(df_prods, on = 'product_id', how = 'left', indicator = True)
```

```python
# This will create a new dataframe, df_merged_large, that contains the combined df_ords and df_ords_prior dataframes and uses the “order_id” column as its key. 
# Only the rows that match on product_id in both dataframes will be kept/included in the resulting dataframe.
df_ords_prods_merge = df_prods.merge(df_ords_products_combined, on = 'product_id', indicator = 'merge_')
```
