```python
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge.pkl'))
```

```python
#creating a subset
df = ords_prods_merge[:1000000]
```

```python
df.head(10)
```

```python
#this function will group a given dataframe by a given column
df.groupby('product_name')
```

```python
#applying the agg() function to each group to obtain the mean values for the “order_number” column
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#grouping without the agg function
df.groupby('department_id')['order_number'].mean()
```

```python
#grouping without the square brackets and using dot notation
df.groupby('department_id').order_number.mean()
```

```python
#aggregate functions mean min max
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
#creating the max_order column while splitting the data based on the user id column and transforming the order_number column for maximum orders for each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
#displays the first 100 rows
ords_prods_merge.head(100)
```

```python
#create a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
