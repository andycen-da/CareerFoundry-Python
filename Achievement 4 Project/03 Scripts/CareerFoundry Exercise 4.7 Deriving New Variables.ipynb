```python
#defining a new function called price_label
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#applying the function to my df
df_ords_prods_million['price_range'] = df_ords_prods_million.apply(price_label, axis=1)
```
